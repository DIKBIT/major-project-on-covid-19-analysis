```
# forecasting of the model using fbprophet
# creating a new dataframe with two coloumns for the forecasting

df_new = df[['Date','new_cases']]
df_new.info()
```

```
# renaming the dataframe
# future dataframe 
df_new = df_new.rename(columns={'Date':'ds','new_cases':'y'})
df_new
```

```
df_new.to_csv('covid analysis',index=False)
```

```
# calling fbprophet 
import fbprophet
op=fbprophet.Prophet(changepoint_prior_scale =5.5)
op.fit(df_new)
```

```
# calling for the future dataframe 
forecast = op.make_future_dataframe(periods = 365, freq='D')
forecast = op.predict(forecast)
```
